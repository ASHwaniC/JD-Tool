```python
from bs4 import BeautifulSoup
import requests

r = requests.get("https://www.naukri.com/project-engineer-jobs-in-bangalore")
c= r.content
soup = BeautifulSoup(c,"html.parser")
page_nr=soup.find_all("div",{"class":"count"})
m=[]
p={}
for item in page_nr:
    try:
        p["page"]=item.find("span",{"class":"cnt"}).text
    except:
        p["page"]=None
    m.append(p)
import pandas as pd
df1=pd.DataFrame(m)
df1['page'] = df1.page.str.split('of').str.get(1)
df0=df1.to_string()
l=[int(s) for s in df0.split() if s.isdigit()]
l2=int(l[1]/50)

l1=[]

for page in range(1,int(l2),1):

    link_set = set()
    response =  requests.get("https://www.naukri.com/project-engineer-jobs-in-bangalore-"+str(page))
    html_data = response.text
    soup = BeautifulSoup(html_data)
    debaters_div = soup.find('div', {"class":"container fl"})
    list_items = debaters_div.find_all("div", {"class":"row "})

    for each_item in list_items:
        html_link = each_item.find('a').get('href')
        link_set.add(html_link)
    link=list(link_set)


    for n in range (0,49,1):
        s=requests.get(link[n])
        e=s.content
        soup = BeautifulSoup(e,"html.parser")

        al=soup.find_all("div",{"class":"sumHd bnr"})
        re=soup.find_all("div",{"class":"sumFoot"})
        jd=soup.find_all("div",{"class":"JD"})
        ed=soup.find_all("div",{"class":"jDisc mt10 edu"})
        
        e={}
        
        link2 = set()
        response =  requests.get(link[n])
        html_data = response.text
        soup = BeautifulSoup(html_data)
        mn=soup.find('div',{"class":"JD"})
        try:
            debaters_div = mn.find('div', {"class":"ksTags"})
        except:
            pass
        try:
            link3=debaters_div.find_all('a',{"target":"_blank"})
        except:
            pass
        for a in link3:
            try:
                link1=a['href']
            except:
                pass
            link2.add(link1)
        try:
            e["Key Skills"]=str(link2)
        except:
            pass
        

        for item in al:
            try:
                e["Title"]=item.find("h1").text
            except:
                e["Title"]= None
            try:
                e["Hiring Organization"]=item.find("a").text
            except:
                e["Hiring Organization"]= None
            try:
                e["Experience"]=item.find("span").text
            except:
                e["Experience"]= None
            try:
                e["Location"]=item.find("div",{"class":"loc collapsed"}).text
            except:
                e["Location"]= None
        for item in re:
            try:
                e["Salary"]=item.find("span",{"class":"sal"}).text
            except:
                e["Salary"]= None
            try:
                e["Job Views"]=item.find("span",{"class":"fr jViews"}).text.replace("Job Views: ","")
            except:
                e["Job Views"]= None
            try:
                e["Job Applicants"]=item.find("span",{"class":"fr jApplys"}).text.replace("Job Applicants: ","")
            except:
                e["Job Applicants"]= None
        for item in jd:
            try:
                e["Job Description"]=item.find("ul").text
            except:
                e["Job Description"]= None
            try:
                e["Ind"]=item.find("div",{"class":"jDisc mt20"}).text
            except:
                e["Ind"]=None
            
            try:
                e["Desired Candidate Profile"]=item.find("ul",{"class":"listing mt15"}).text
            except:
                e["Desired Candidate Profile"]= None
            try:
                e["Company Profile"]=item.find("div",{"class":"mt15 lH20 companyProfile" }).text
            except:
                e["Company Profile"]=None
            try:
                e["Education"]=item.find("div",{"class":"jDisc mt10 edu"}).text
            except:
                e["Education"]= None
            l1.append(e)

        import pandas as pd
        df = pd.DataFrame(l1)

        
        
df
```

```python
try:
    df['Min Experience'] = df.Experience.str.split('-').str.get(0)
except:
    df['Min Experience'] = None
try:
    df['Max Experience'] = df.Experience.str.split('-').str.get(1) 
except:
    df['Max Experience'] = None
try:
    df['Min Salary'] = df.Salary.str.split('-').str.get(0)
except:
    df['Min Salary'] = None
try:
    df['Max Salary'] = df.Salary.str.split('-').str.get(1)
except:
    df['Max Salary'] = None
try:
    df['UG'] = df.Education.str.split(':').str.get(1)
except:
    df['UG'] = None
try:
    df['PG'] = df.Education.str.split(':').str.get(2)
except:
    df['PG'] = None
try:
    df['Doctorate'] = df.Education.str.split(':').str.get(3)
except:
    df['Doctorate'] = None
try:
    df['Industry'] = df.Ind.str.split(':').str.get(2)
except:
    df['Industry'] = None
try:
    df['Functional Area'] = df.Ind.str.split(':').str.get(3)
except:
    df['Functional Area'] = None
try:
    df['Role Category'] = df.Ind.str.split(':').str.get(4)
except:
    df['Role Category'] = None
try:
    df['Role'] = df.Ind.str.split(':').str.get(5)
except:
    df['Role'] = None
try:
    df['Employment Type'] = df.Ind.str.split(':').str.get(6)
except:
    df['Employment Type'] = None

df['PG']=df.PG.str.replace('Doctorate','')
df['UG']=df['UG'].str.replace('PG','')
df['Industry']=df['Industry'].str.replace('Functional Area','')
df['Functional Area']=df['Functional Area'].str.replace('Role Category','')
df['Role Category']=df['Role Category'].str.replace('Role','')
df['Role']=df['Role'].str.replace('Employment Type','')

del df['Salary']
del df['Experience']
del df['Education']
del df['Ind']

df['Key Skills']=df['Key Skills'].str.replace('jobs','')
df['Key Skills']=df['Key Skills'].str.replace('-',' ')
df['Key Skills']=df['Key Skills'].str.replace('{','')
df['Key Skills']=df['Key Skills'].str.replace('}','')
df['Key Skills']=df['Key Skills'].str.replace("'",'')
df['Max Salary']=df['Max Salary'].str.replace("P.A.",'')
df['Max Experience']=df['Max Experience'].str.replace("yrs",'')
df
```
